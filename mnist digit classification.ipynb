# **Getting the data from Keras**

```python
objetcs = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = objetcs.load_data()
```

# **Visualizing the data**

```python
print(X_train.shape)
print(X_train[0])
```

```python
# normalizing the data
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
for i in range(9):
  # define subplot
  plt.subplot(330+1+i)

  # plot raw pixel data
  plt.imshow(X_train[i])
```

# **Building the Model**

```python
model = tf.keras.models.Sequential([
     tf.keras.layers.Flatten(input_shape=(28,28)),  # flatten makes the LOL to list
     tf.keras.layers.Dense(128,activation='relu'),
     tf.keras.layers.Dense(10,activation=tf.nn.softmax)
    ])
```

# **Optimizer and Cost Function of the Model**

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=25)
```

# **Testing the Model**

```python
print(model.evaluate(X_test, y_test))
```

# **Prediction**

```python
plt.imshow(X_test[0])
prediction = model.predict(X_test)
print(np.argmax(prediction[0]))
```
